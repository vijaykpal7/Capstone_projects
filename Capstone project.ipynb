```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import service
from selenium.webdriver.chrome.options import Options
import time
import random
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.olx.in/en-in/uttaranchal_g2001175/for-sale-houses-apartments_c1725?filter=listed_by_eq_owner%2Ctype_eq_houses")
```

```python
link=[]
screen_height = driver.execute_script("return window.innerHeight")
A = 1

while True:
    # Scroll down by one screen height
    driver.execute_script(f"window.scrollTo(0, {screen_height * A});")
    A += 1
    time.sleep(2)
    
    # Update scroll height each time after scrolling
    scroll_height = driver.execute_script("return document.body.scrollHeight")
    
    # Break the loop if the height we need to scroll to is larger than the total scroll height
    if screen_height * A > scroll_height:
        try:
            load_more = driver.find_element(By.XPATH, "//*[contains(text(),'load more')]")
            # Scroll to the load more button
            driver.execute_script("arguments[0].scrollIntoView(true);", load_more)
            time.sleep(2)
            # Click the load more button using JavaScript
            driver.execute_script("arguments[0].click();", load_more)
            time.sleep(2)
            A = 1  # Reset the counter after clicking load more
        except Exception as e:
            print("No more 'load more' button found or an error occurred:", e)
            break

soup = BeautifulSoup(driver.page_source, 'html.parser')
items = soup.find_all('a')
hrefs = [link.get('href') for link in items]

for href in hrefs:
    if 'for-sale-houses-apartments' and 'bhk-houses' in href:
        link.append(href)
```

```python
bhk_type=[]
property_type=[]
beds=[]
baths=[]
balcony=[]
address=[]
city=[]
state_code=[]
country=[]
landmark=[]

carpetarea=[]
status=[]
floor=[]
transaction_type=[]
year_built=[]
price_per_sqft=[]

furnished_status=[]
facing=[]
car_parking=[]
type_of_ownership=[]
booking_amount=[]
buy_total_price=[]


count = 0
for i in new_link:
    driver.get(i)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    count+=1
    print(count,end=" ")
    time.sleep(3)


    try:
        f=driver.find_element(By.CSS_SELECTOR,"#main_content > div > div > div > div.rui-jIycK.rui-tOTz6.rui-EY2Ik.rui-RVnVw.rui-ZvM8x.rui-VAm0G > div.rui-99hme._7qT_t > div > section > div > div._3ipDN > div > span")
        address.append(f.text)
    except:
        address.append(None)

    try:
        span_city = soup.find('span', class_='_1RkZP')

        text_parts = span_city.get_text().split(', ')
        city.append(text_parts[1]) 
        state_code.append(text_parts[2] )
    except:
        city.append(None) 
        state_code.append(None)
    
    country.append('India')

    
    try:
        # Find the location
        location = None
        for p in soup.find_all('p'):
            text = p.get_text().strip().lower()
            if text.startswith('location'):
                location = text.split(' - ')[1]  # Split the text to get the location part
                break
    
        # Append null if location is not found
        if location is None:
            landmark.append(None)
        else:
            landmark.append(location)
    
    except Exception as e:
        # In case of any error, append null
        landmark.append(None)

    found_values = {}
    
    try:
        item_params_div = soup.find('div', {'data-aut-id': 'itemParams'})
        items = item_params_div.find_all('div', class_='_1O2tT')
        
        # Extract key-value pairs
        for item in items:
            key = item.find('span', {'class': '_3V4pD'}).text.strip()
            value = item.find('span', {'class': 'B6X7c'}).text.strip()
            found_values[key] = value
        
        # Append values or None if not found
        property_type.append(found_values.get('Type', None))
        bhk_type.append(found_values.get('Bedrooms', None))
        beds.append(found_values.get('Bedrooms', None))
        baths.append(found_values.get('Bathrooms', None))
        furnished_status.append(found_values.get('Furnishing', None))
        status.append(found_values.get('Construction Status', None))
        carpetarea.append(found_values.get('Carpet Area (ft²)', None))
        car_parking.append(found_values.get('Car Parking', None))
        facing.append(found_values.get('Facing', None))
        floor.append(found_values.get('Total Floors', None))
        
    except AttributeError as e:
        print(f"An error occurred: {e}")
        # Append None for each list if an error occurs
        property_type.append(None)
        bhk_type.append(None)
        beds.append(None)
        baths.append(None)
        furnished_status.append(None)
        status.append(None)
        carpetarea.append(None)
        car_parking.append(None)
        facing.append(None)
        floor.append(None)
    
    try:
        h=driver.find_element(By.CSS_SELECTOR,"#main_content > div > div > div > div.rui-jIycK.rui-tOTz6.rui-EY2Ik.rui-RVnVw.rui-ZvM8x.rui-VAm0G > div.rui-99hme._7qT_t > div > section > span.T8y-z")
        buy_total_price.append(h.text)
    except:
        buy_total_price.append(None)
    transaction_type.append(None)
    year_built.append(None)
    type_of_ownership.append(None)
    booking_amount.append(None)
    balcony.append(None)
```

```python
df=pd.DataFrame({'bhk_type':bhk_type,'property_type':property_type,'beds':beds,'baths':baths,'balcony':balcony,'address':address,'city':city,
                    'state_code':state_code,
                    'country':country,
                    'landmark':landmark,
                    'carpetarea':carpetarea,
                    'status':status,
                    'floor':floor,
                    'transaction_type':transaction_type,
                    'year_built':year_built,
                    # 'price_per_sqft':price_per_sqft,
                    'furnished_status':furnished_status,
                    'facing':facing,
                    'car_parking':car_parking,
                    'type_of_ownership':type_of_ownership,
                    'booking_amount':booking_amount,
                    'buy_total_price':buy_total_price})
```

```python
df
```

```python
df['buy_total_price'] = df['buy_total_price'].replace('[\₹,]', '', regex=True).astype(float)
df['carpetarea'] = df['carpetarea'].replace('[\₹,]', '', regex=True).astype(float)

# Calculate the 'price_per_sqft'
df['price_per_sqft'] = df['buy_total_price'] / df['carpetarea']

# Round the 'price_per_sqft' to the desired precision, for example, 2 decimal places
df['price_per_sqft'] = df['price_per_sqft'].round(0)
```

```python
df.to_csv('D:\finalpropertydata.csv')
```
